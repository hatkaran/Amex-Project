```python
import pandas as pd
import numpy as np

# Define a function to replace outliers with column median using Z-score
def replace_outliers_with_median_zscore(df, z_thresh=3):

    for column in df.columns:
        non_null_mask = df[column].notnull()  # Create a boolean mask for non-null values

        column_median = df.loc[non_null_mask, column].median()  # Calculate median using non-null values
        column_std = df.loc[non_null_mask, column].std()  # Calculate standard deviation using non-null values

        z_scores = np.abs((df[column] - column_median) / column_std)
        df[column] = np.where(z_scores > z_thresh, np.where(non_null_mask, column_median, np.nan), df[column])

    return df

# Call the function to replace outliers with median
Eval[["merchant_spend_01","merchant_spend_02","merchant_spend_04","merchant_spend_05","merchant_spend_06","merchant_spend_07","merchant_spend_08","customer_spend_01","customer_spend_18"]] = replace_outliers_with_median_zscore(test.loc[:,["merchant_spend_01","merchant_spend_02","merchant_spend_04","merchant_spend_05","merchant_spend_06","merchant_spend_07","merchant_spend_08","customer_spend_01","customer_spend_18"]])

Eval
```
